```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/DALI/Sample - Superstore.csv', index_col=0)
df = df.drop(['Row ID'], axis=1)
import pdb
```

```python
import pandas as pd

def fill_missing_values1(df):

    # Create a dictionary to map names to IDs and vice versa
    name_dict = {}
    id_dict = {}
    unique = 0
    for index, row in df.iterrows():
      name = row['Customer Name']
      id = row['Customer ID']
      #collect all the matching pairs and assign sequential number
      if not pd.isna(name) and not pd.isna(id):
        if not id in id_dict and not name in name_dict:
          unique = unique + 1
          name_dict[name]=unique
          id_dict[id]=unique
    #listing of sequential order
    sorted_name = sorted(name_dict.items(), key=lambda x: x[1])
    ordered_name = [item[0] for item in sorted_name]
    sorted_id = sorted(id_dict.items(), key=lambda x: x[1])
    ordered_id = [item[0] for item in sorted_id]

    #fill in the missing info for known pairs
    for index, row in df.iterrows():
      name = row['Customer Name']
      id = row['Customer ID']
      if pd.isna(name) and id in id_dict:
        df.at[index, 'Customer Name'] = ordered_name[id_dict[id]-1]
      if pd.isna(id) and name in name_dict:
        df.at[index, 'Customer ID'] = ordered_id[name_dict[name]-1]

    #redirect ID to be our assigned from 1,2,...n
    for index, row in df.iterrows():
      name = row['Customer Name']
      id = row['Customer ID']
      if not pd.isna(id) and id in id_dict:
        df.at[index, 'Customer ID'] = id_dict[id]
      if not pd.isna(name) and name in name_dict:
        df.at[index, 'Customer ID'] = name_dict[name]
    return df

def fill_missing_values2(df):
    #Same process for product name/id
    name_dict = {}
    id_dict = {}
    unique = 0
    for index, row in df.iterrows():
      name = row['Product Name']
      id = row['Product ID']
      if not pd.isna(name) and not pd.isna(id):
        if not id in id_dict and not name in name_dict:
          unique = unique + 1
          name_dict[name]=unique
          id_dict[id]=unique
    sorted_name = sorted(name_dict.items(), key=lambda x: x[1])
    ordered_name = [item[0] for item in sorted_name]
    sorted_id = sorted(id_dict.items(), key=lambda x: x[1])
    ordered_id = [item[0] for item in sorted_id]

    for index, row in df.iterrows():
      name = row['Product Name']
      id = row['Product ID']
      if pd.isna(name) and id in id_dict:
        df.at[index, 'Product Name'] = ordered_name[id_dict[id]-1]
      if pd.isna(id) and name in name_dict:
        df.at[index, 'Product ID'] = ordered_id[name_dict[name]-1]

    for index, row in df.iterrows():
      name = row['Product Name']
      id = row['Product ID']
      if not pd.isna(id) and id in id_dict:
        df.at[index, 'Product ID'] = id_dict[id]
      if not pd.isna(name) and name in name_dict:
        df.at[index, 'Product ID'] = name_dict[name]

    return df

def cut(df):
    #if we don't know either name or id are connected, then ignore- this is only approximately 200 entries
    missingIdx = df[pd.isna(df['Customer Name']) | pd.isna(df['Customer ID']) | pd.isna(df['Product Name']) | pd.isna(df['Product ID'])].index
    df.drop(missingIdx, inplace=True)
    return df
```

```python
df = fill_missing_values1(df)
df = fill_missing_values2(df)
df = cut(df)

df.to_csv('/content/drive/MyDrive/DALI/Superstore2.csv')
df
```
